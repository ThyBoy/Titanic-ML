```python
trdata=data[["Pclass","Age","Fare","Sex","Embarked","SibSp","Parch"]]
trdata.head()
```

```python
trdata.isnull().sum()
```

```python
trdata.Embarked.value_counts()
```

```python
trdata.fillna({'Age':trdata.Age.mean(),
           'Embarked':'S'},inplace=True)
```

```python
trdata.isnull().sum()
```

```python
data1=pd.get_dummies(trdata[['Sex','Embarked']])
data1.head()
```

```python
def aply(row):
    if (row.SibSp==0) & (row.Parch==0):
        return(0)
    elif (row.SibSp!=0) | (row.Parch!=0):
        return(1)
    else:
        print("incorrect")
```

```python
data2=trdata[["Age","Fare"]]
data3=trdata[["Pclass"]]
data4=pd.get_dummies(data3.astype(str))
data4["Relatives"]=trdata.apply(lambda row: aply(row),axis=1)
data4.head()
```

```python
train_data=pd.concat((data1,data2,data4),axis=1)
train_data.Fare.sort_values(ascending=False)
train_data.head()
```

```python
output=data[["Survived"]]
output.head()
```

```python
X=train_data.iloc[:,:].values
Y=output.iloc[::].values
print(X[0:3])
```

```python
from sklearn.preprocessing import StandardScaler
age_scaler=StandardScaler()
X[:,5:6]=age_scaler.fit_transform(X[:,5:6])
fare_scaler=StandardScaler()
X[:,6:7]=age_scaler.fit_transform(X[:,6:7])
print(X[0:3])
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size=.20,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(xtrain,ytrain)
```

```python
ypred=log.predict(xtest)

from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.metrics import plot_confusion_matrix

plot_confusion_matrix(log,xtest,ytest)
acc=accuracy_score(ypred,ytest)
print(acc)

cm=confusion_matrix(ytest,ypred)
print(cm)
print("\n\n")
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=11)
knn.fit(xtrain,ytrain)
y_pred=knn.predict(xtest)

plot_confusion_matrix(knn,xtest,ytest)
acc=accuracy_score(y_pred,ytest)
print(acc)

cm=confusion_matrix(ytest,y_pred)
print(cm)
print("\n\n")
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=30)
rfc.fit(xtrain,ytrain)
ypred=rfc.predict(xtest)

plot_confusion_matrix(rfc,xtest,ytest)
acc=accuracy_score(ypred,ytest)
print(acc)

cm=confusion_matrix(ytest,ypred)
print(cm)
print("\n\n")
```

```python
from sklearn.svm import SVC
svc=SVC(C=0.2,kernel='rbf',degree=3)
svc.fit(xtrain,ytrain)
ypred=svc.predict(xtest)

plot_confusion_matrix(svc,xtest,ytest)
acc=accuracy_score(ypred,ytest)
print(acc)

cm=confusion_matrix(ytest,ypred)
print(cm)
print('\n\n')
```
